# Box Plot

Prompt : Beri kode untuk menampilkan Box Plot dari dataset

```python
import matplotlib.pyplot as plt

# Membuat Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot([dataset[dataset['2urvived'] == 0]['Age'].dropna(),
             dataset[dataset['2urvived'] == 1]['Age'].dropna()],
            labels=['Not Survived', 'Survived'])
plt.title('Box Plot of Age by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

# Scatter Plot

Prompt : Beri kode program untuk menampilkan scatter plot dari dataset

```python
# Membuat Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(dataset['Age'], dataset['Fare'], c=dataset['2urvived'], cmap='coolwarm', alpha=0.6)
plt.title('Scatter Plot of Age vs Fare (Survival Status)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar(label='2urvived')
plt.grid(True)
plt.show()
```

# Heat Map

Prompt: Beri kode program untuk menampilkan Heat Map dari dataset

```python
import seaborn as sns

# Mengisi nilai yang hilang dengan nilai rata-rata untuk fitur 'Age'
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)

# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = dataset.corr()

# Membuat Heatmap
plt.figure(figsize=(20, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
