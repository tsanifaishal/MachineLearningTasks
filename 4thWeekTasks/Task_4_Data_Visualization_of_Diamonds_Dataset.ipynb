# Feature Statistic

Prompt : beri kode untuk menampilkan feature statistic dari dataset diamonds.csv

```python
# Menampilkan statistik fitur
print("Statistik Fitur Dataset Diamonds:")
print(dataset.describe())

```

# Box Plot

Prompt : Beri kode untuk menampilkan Box Plot dari dataset

```python
import matplotlib.pyplot as plt

# Menampilkan boxplot
dataset.drop(columns=['Unnamed: 0']).boxplot(figsize=(10, 6))
plt.title('Boxplot untuk Dataset')
plt.xticks(rotation=45)
plt.show()
```

#Scatter Plot

Prompt : Buatkan kode program untuk menampilkan scatter plot dari dataset

```python
# Menampilkan scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(dataset['carat'], dataset['price'], alpha=0.5)
plt.title('Scatter Plot antara Carat dan Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

# Heat Map

Prompt : Buatkan kode program untuk menampilkan Heat Map dari dataset

```python
import seaborn as sns

# Menghitung korelasi antar fitur
correlation_matrix = dataset.corr()

# Menampilkan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antar Fitur dalam Dataset')
plt.show()
```
